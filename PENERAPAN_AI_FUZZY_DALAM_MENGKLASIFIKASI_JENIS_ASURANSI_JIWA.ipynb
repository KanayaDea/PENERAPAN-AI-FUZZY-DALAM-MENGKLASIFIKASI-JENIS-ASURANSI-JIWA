Proses Fuzzifikasi, menentukan fungsi keanggotaan lalu menampilkan grafiknya dengan matplotlip

```python
import matplotlib.pyplot as plt #Fungsi Keanggotaan untuk gaji Tinggi
import numpy as np

# Definisi fungsi keanggotaan
def gaji_tinggi(x):
    if x > 30:
        return 1
    elif x < 20 :
        return 0
    elif x > 20 and x <= 30:
        return (x - 20) / (30 - 20)
    else:
        return 0

# Domain
x = np.linspace(0, 40, 100)  # Sesuaikan dengan domain yang sesuai

# Hitung nilai keanggotaan
y_rendah = [gaji_tinggi(i) for i in x]

# Buat grafik
plt.plot(x, y_rendah)
plt.title('Fungsi Keanggotaan untuk gaji Tinggi')
plt.xlabel('Gaji Tinggi (Rp. juta)')
plt.ylabel('Nilai Keanggotaan')
plt.grid(axis='y')
plt.legend()
plt.show()
```

Plot gabungan dari fungsi keanggotaan tingkat gaji

```python
import matplotlib.pyplot as plt
import numpy as np

# Definisi fungsi keanggotaan
def gaji_tinggi(x):
    if x > 30:
        return 1
    elif x < 20 :
        return 0
    elif x > 20 and x <= 30:
        return (x - 20) / (30 - 20)
    else:
        return 0

def gaji_rendah(x):
    if x <= 4 :
        return 1
    elif x > 5 :
        return 0
    elif x >= 4 and x < 5:
        return (5 - x) / (5 - 4)
    else:
        return 0

def gaji_menengah(x):
    if x > 4 and x < 5 :
        return (x - 4)/(5 - 4)
    elif x < 4 and x > 21 :
        return 0
    elif x >= 5 and x <= 20:
        return 1
    elif x >= 20 and x <= 21:
        return (21 - x) / (21 - 20)
    else:
        return 0

# Domain
x = np.linspace(0, 35, 100)  # Sesuaikan dengan domain yang sesuai

# Hitung nilai keanggotaan untuk masing-masing fungsi
y_tinggi = [gaji_tinggi(i) for i in x]
y_rendah = [gaji_rendah(i) for i in x]
y_menengah = [gaji_menengah(i) for i in x]

# Buat grafik
plt.plot(x, y_tinggi, label='Gaji Tinggi')
plt.plot(x, y_rendah, label='Gaji Rendah')
plt.plot(x, y_menengah, label='Gaji Menengah')
plt.title('Fungsi Keanggotaan Tingkat Gaji')
plt.xlabel('Gaji(Juta)')
plt.ylabel('Nilai Keanggotaan')
plt.grid(axis='y')
plt.legend()
plt.show()
```

Plot gabungan untuk fungsi keanggotaan riwayat kesehatan

```python
import matplotlib.pyplot as plt
import numpy as np

# Definisi fungsi keanggotaan
def riwayat_kronis(x):
    if x > 5:
        return 1
    elif x < 3 :
        return 0
    elif x > 3 and x <= 5:
        return (x - 3) / (5 - 3)
    else:
        return 0

def riwayat_sehat(x):
    if x <= 1 :
        return 1
    elif x > 5 :
        return 0
    elif x > 1 and x <= 5:
        return (5 - x) / (5 - 1)
    else:
        return 0

def riwayat_sedang(x):
    if x > 1 and x <= 2 :
        return (x - 1)/(2 - 1)
    elif x <= 1 and x > 5 :
        return 0
    elif x >= 2 and x <= 4:
        return 1
    elif x > 4 and x <= 5:
        return (5 - x) / (5 - 4)
    else:
        return 0

# Domain
x = np.linspace(0, 9, 100)  # Sesuaikan dengan domain yang sesuai

# Hitung nilai keanggotaan untuk masing-masing fungsi
y_kronis = [riwayat_kronis(i) for i in x]
y_sehat = [riwayat_sehat(i) for i in x]
y_sedang = [riwayat_sedang(i) for i in x]

# Buat grafik
plt.plot(x, y_kronis, label='Kronis')
plt.plot(x, y_sehat, label='Sehat')
plt.plot(x, y_sedang, label='Sedang')
plt.title('Fungsi Keanggotaan Riwayat Penyakit')
plt.xlabel('Jumlah penyakit')
plt.ylabel('Nilai Keanggotaan')
plt.grid(axis='y')
plt.legend()
plt.show()
```
